```python
import numpy as np
```

```python
def Trapezoidal(f, a, b):
    answ = []
    h_arr = []
    for k in range(1,5):
        n = 2**k + 1
        h = (b-a)/float(n)
        s = 0.5*(f(a) + f(b))
        for i in range(1,n,1):
            s = s + f(a + i*h)
        answ.append(h*s)
        h_arr.append(h)
        
    return answ, h_arr

def f(x):
    return x**(3/5)

a = 0
b = 1

result, h_arr = Trapezoidal(f, a, b)
exact1 = 0.625
err1 = abs(exact1-np.asarray(result))
```

```python
def f2(x):
    return x**(8/5)
result2, h_arr2 = Trapezoidal(f2, a, b)
exact2 = 0.3846153846153846
err2 = abs(exact2-np.asarray(result2))

def f3(x):
    return x**(13/5)
result3, h_arr3 = Trapezoidal(f3, a, b)

exact3 = 0.2777777777777778
err3 = abs(exact3-np.asarray(result3))
```

```python
import pandas as pd
df_trapez = pd.DataFrame({'Values': result,'Error': err1})
df_trapez2 = pd.DataFrame({'Values': result2,'Error': err2})
df_trapez3 = pd.DataFrame({'Values': result3,'Error': err3})
```

```python
def simpson(a, b, f):
    sum = 0
    ans = []
    for m in range(1,5):
        sum = 0
        n = 2**m 
        inc = (b - a) / n
        for k in range(n+1):
            x = a + (k * inc)
            summand = f(x)
            if (k != 0) and (k != n):
                summand *= (2 + (2 * (k % 2)))
            sum += summand
        ans.append(((b - a) / (3 * n)) * sum)
    return ans
```

```python
a = 0
b=  1
result_simp = simpson(a,b,f)
result_simp2 = simpson(a,b,f2)
result_simp3 = simpson(a,b,f3)
```
