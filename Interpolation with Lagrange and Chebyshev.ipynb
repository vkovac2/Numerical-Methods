```python
import numpy as np
import matplotlib.pyplot as plt

a =-6
b =6
n=21

def lagrange(points, i):
    s = ''
    for j in range(len(points)):
        if j == i:
            continue
        s += ('('+'x'+'-'+ str(points[j])+')'+'/'+'('+str(points[i]) +'-'+ str(points[j])+')*')
        
    
    return s[:-1]

def interpolate(points, f):
    s = ''
    for i, x in enumerate(points):  
        s += str(f(x)) + '*' + lagrange(points, i) +'+'
        
    return lambda x : eval(s[:-1])
    

def f(x):
    return 1/(x**2 + 1)


f_lagrange = interpolate(np.linspace(a, b, n), f)

z_values = []
for i in range(1,n+2):
    z_values.append(np.cos((2*i/1)/(2*n+2)*np.pi))
#     z values

x_cheb = []
for i in range(0,n+1):
    x_cheb.append(0.5*(a+b+(b-a)*z_values[i]))
    #Chebyshev points
    
f_chebyshev = interpolate(x_cheb, f) 


x = np.linspace(a, b, 500)
# plt.xlim(-6,6)
# plt.ylim()
y_lagrange = f_lagrange(x)
y_chebyshev = f_chebyshev(x)
ytrue = f(x)
plt.ylim(-10,10)
plt.plot(x, ytrue,'r')
plt.plot(x, y_lagrange,'b')
plt.plot(x, y_chebyshev,'m')
plt.show()
```
